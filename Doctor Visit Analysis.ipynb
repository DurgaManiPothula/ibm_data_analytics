# Visualize the percentage of people having govt & private health insurance

```python
#govt insurance due to low income

label =['yes', 'no']

Y= df[df['freepoor']=='yes']
N= df[df['freepoor']== 'no']

x=[ Y.shape[0], N.shape[0]]

plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people getting govt health insurance due to low income")
plt.show()
```

```python
#govt insurance due to old age

#label =['yes', 'no']

Y= df[df['freerepat']=='yes']
N= df[df['freerepat']== 'no']

x=[ Y.shape[0], N.shape[0]]

plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people getting govt health insurance due to old age")
plt.show()
```

```python
#private insurance

#label =['yes', 'no']

Y= df[df['private']=='yes']
N= df[df['private']== 'no']

x=[ Y.shape[0], N.shape[0]]

plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people getting private health insurance")
plt.show()
```

# Plot a bar chart to analyze the no of reduced days of activity due to illness

```python
db = df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'], db['reduced'], color =['cornflowerblue', 'lightseagreen'])
plt.title('Bar Chart')
plt.xlabel('Reduced Activity')
plt.ylabel('Gender')
plt.show()
```
